### 함수를 이용한 그래프 구현 
#### 그래프만 볼 수 있을 뿐 다른 정보는 볼 수 없다. ex) y, A, f 등등 

```python
def draw_sin(t, A, f, b):
    y = A * np.sin(np.pi * f * t) + b

    plt.figure(figsize=(12, 6))
    plt.plot(t, y)
    plt.grid()
    plt.show()
```

```python
t = np.arange(0, 5, 0.01)
draw_sin(t, 1, 1, 0)
```

```python
# 클래스 선언 
class DrawSin:
    # 던더, 더블 언더 스코어 : '__'
    # 아래의 던더 init은 매개변수 초기화한다.
    # self는 생선된 인스턴스를 가리키는 매개변수다. 
    # amp, freq, bias, end_time은 클래스 생성자에서 사용되는 매개변수로, 인스턴스를 생성할 때 전달되는 값이다. 

    # 즉, 이 코드는 DrawSin 클래스의 생성자를 정의하는 것으로, amp, freq, bias, end_time 값을 인자로 받아서 클래스의 인스턴스를 초기화 한다.

    def __init__(self, amp, freq, bias, end_time):
        self.amp = amp
        self.freq = freq
        self.bias = bias
        self.end_time = end_time
```

#### 삼각함수 그래프 클래스 

```python
# 클래스 선언 
class DrawSin:
    def __init__(self, amp, freq, bias, end_time):
        self.amp = amp
        self.freq = freq
        self.bias = bias
        self.end_time = end_time
    # 함수계산 
    def calc_sin(self):
        # 정의역 생성 
        self.t = np.arange(0, self.end_time, 0.01)
        return self.amp * np.sin(2 * np.pi * self.freq * self.t) + self.bias
    # 그래프 생성 
    def draw_sin(self):
        # 치역 생성 
        y = self.calc_sin()    
        plt.figure(figsize=(12, 6))
        plt.plot(self.t, y)
        plt.grid()
        plt.show()
```

```python
ds = DrawSin(1,1,0,7)
ds.draw_sin()
ds.t
```

```python
ds.freq = 2
ds.draw_sin()
```

#### 메서드 오버라이딩 (덮어쓰기)

```python
# 기존 클래스DS를 상속받는 DS2 클래스 생성
# 즉, 기존 draw_sin 함수를 새로운 기능을 추가하여 덮어쓰기 함 
class DrawSinusoidal2(DrawSinusoidal):
    def draw_sin(self):
        y = self.calc_sin()
        plt.figure(figsize=(12, 6))
        plt.plot(self.t, y)
        plt.title('Sin Graph ')
        plt.ylabel('Sin')
        plt.xlabel('time (sec)')
        plt.grid()
        plt.show()
```

```python
dc2 = DrawSinusoidal2(1, 1, 0, 7)
dc2.draw_sin()
```

#### 클래스에서 super()의 사용 

```python
# 오버 라이딩 발생 클래스가 재 정의됨 (덮어 쓴 것이 아닌 상위 매소드를 가리게 된 것)
class DrawSinusoidal3(DrawSinusoidal2):
    def __init__(self, amp, freq, bias, end_time, ts):
        self.ts = ts
```

##### 오버라이딩을 방지하고, 기존 속성을 상속받으려면 super() 사용 

```python
class DrawSinusoidal3(DrawSinusoidal2):
    def __init__(self, amp, freq, bias, end_time, ts):
        super().__init__(amp, freq, bias, end_time)
        self.ts = ts
```

```python
ds3 = DrawSinusoidal3(1,1,0,7,0.01)
ds3.__dict__
```

```python
class DrawSinusoidal3(DrawSinusoidal2):
    def __init__(self, amp, freq, bias, end_time, ts):
        super().__init__(amp, freq, bias, end_time)
        self.ts = ts
        
    def calc_sin(self):
        self.t = np.arange(0, self.end_time, self.ts)
        return self.amp * np.sin(2 * np.pi * self.freq * self.t) + self.bias
    def draw_sin(self):
        y = self.calc_sin()
        plt.figure(figsize=(12, 6))
        plt.plot(self.t, y)
        plt.title('Sin Graph ')
        plt.ylabel('Sin')
        plt.xlabel('time (sec)')
        plt.grid()
        plt.show()
```

```python
ds3 = DrawSinusoidal3(1,1,0,7,0.1)
ds3.draw_sin()
```
